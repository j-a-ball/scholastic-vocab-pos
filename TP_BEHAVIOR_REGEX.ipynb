```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import spacy
import pandas as pd
import re
nlp = spacy.load('en_core_web_lg')
```

```python
tpb = pd.read_csv('data/TP_behavior.csv')
tpb = tpb.rename(columns={'Text (To be Labeled)':'Text'})
tpb = tpb[tpb['Label'] != 'other']
print(tpb.shape)
```

```python
def wordcount(text_list):
    wordcounts = {}
    for text in text_list:
        doc = nlp(text)
        for token in doc:
            if not token.is_stop or token.is_punct:
                if token.lemma_ in wordcounts:
                    wordcounts[token.lemma_] += 1
                else:
                    wordcounts[token.lemma_] = 1
    return wordcounts
```

```python
wordcounts = wordcount(tpb['Text'].tolist())
```

```python
wordcounts = {k: v for k, v in sorted(wordcounts.items(), reverse=True, key=lambda item: item[1])}
```

```python
for pair in list(wordcounts.items())[:100]:
    print(pair)
```
